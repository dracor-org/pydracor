# Extract DraCor character names

```python
# DraCorAPI setup
import requests
import json

# Function to get a corpus by name
def getCorpusData(corpusname):
    requesturl = "https://dracor.org/api/corpus/" + corpusname
    r = requests.get(requesturl)
    corpus = json.loads(r.text)
    return corpus

# Function to get names of plays in a corpus
def getPlaynamesInCorpus(corpusname):
    corpus = getCorpusData(corpusname)["dramas"]
    playnames = []
    for drama in corpus:
        playname = drama["name"]
        playnames.append(playname)
    return playnames

```

```python
def getCastOfDramaInCorpus(playname, corpusname):
    r = requests.get("https://dracor.org/api/corpus/" + corpusname + "/play/" + playname + "/cast")
    cast = json.loads(r.text)
    return cast
```

```python
def getCorporaIds():
    
    r = requests.get("https://dracor.org/api/corpora")
    corpora = json.loads(r.text)
    ids = []
    for corpus in corpora:
        corpusname = corpus["name"]
        ids.append(corpusname)
    return ids
```
